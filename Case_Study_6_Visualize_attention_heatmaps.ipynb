# Lab 6: Attention Heatmap Visualization

## Objective
To visualize how a model focuses on different words using the Attention Mechanism.

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sentence = ["I", "love", "deep", "learning"]
attention_weights = torch.tensor([0.1, 0.3, 0.4, 0.2])

print("Original attention weights:", attention_weights)
print("Sentence:", sentence)
```

```python
attention_weights_normalized = F.softmax(attention_weights, dim=0).detach().numpy()

print("Normalized attention weights:", attention_weights_normalized)

# Plot heatmap
plt.figure(figsize=(8, 2))
sns.heatmap([attention_weights_normalized], annot=True, xticklabels=sentence, cmap="Blues", cbar=True)
plt.title("Attention Heatmap")
plt.xlabel("Words")
plt.ylabel("Attention")
plt.show()
```

## Viva Questions & Answers

**Q1. What is the Attention Mechanism?**

**Q3. What does darker color indicate?**
