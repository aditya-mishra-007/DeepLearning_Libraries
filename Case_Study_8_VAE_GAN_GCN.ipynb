# Lab 8: VAE, GAN, GCN

This lab covers three advanced deep learning architectures: Variational Autoencoder (VAE), Generative Adversarial Network (GAN), and Graph Convolutional Network (GCN).

## 1. Variational Autoencoder (VAE)

### Objective
To generate data using a Variational Autoencoder, a probabilistic generative model.

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

input_dim = 20
latent_dim = 2

inputs = Input(shape=(input_dim,))
encoded = Dense(10, activation='relu')(inputs)
latent = Dense(latent_dim)(encoded)

decoded = Dense(10, activation='relu')(latent)
outputs = Dense(input_dim, activation='sigmoid')(decoded)

vae = Model(inputs, outputs)
vae.compile(optimizer='adam', loss='mse')

print("VAE model summary:")
vae.summary()
```

### Viva Questions for VAE

**Q1. What is Variational Autoencoder?**  
A probabilistic generative model that learns to encode and decode data.

**Q2. What is latent space?**  
Compressed feature representation in lower dimensions.

## 2. Generative Adversarial Network (GAN)

### Objective
To generate realistic data using a Generative Adversarial Network with generator vs discriminator.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

generator = Sequential([
    Dense(16, activation='relu', input_dim=10),
    Dense(1, activation='sigmoid')
])

discriminator = Sequential([
    Dense(16, activation='relu', input_dim=1),
    Dense(1, activation='sigmoid')
])

print("Generator model:")
generator.summary()
print("\nDiscriminator model:")
discriminator.summary()
```

### Viva Questions for GAN

**Q1. What is Generative Adversarial Network?**  
A framework where generator and discriminator compete in training.

**Q2. Role of generator?**  
Creates fake data to fool the discriminator.

**Q3. Role of discriminator?**  
Detects whether data is fake or real.

## 3. Graph Convolutional Network (GCN)

### Objective
To perform graph-based learning using a Graph Convolutional Network for node classification.

```python
import torch
import torch.nn as nn

class SimpleGCN(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = nn.Linear(10, 2)

    def forward(self, x, adj):
        x = torch.matmul(adj, x)
        x = self.fc(x)
        return x

# Example usage
model = SimpleGCN()
print("GCN model:")
print(model)

# Dummy data
x = torch.randn(5, 10)  # 5 nodes, 10 features
adj = torch.randn(5, 5)  # adjacency matrix
output = model(x, adj)
print(f"Input shape: {x.shape}, Output shape: {output.shape}")
```

### Viva Questions for GCN

**Q1. What is Graph Convolutional Network?**  
A neural network designed for graph-structured data.

**Q2. What are nodes and edges?**  
Nodes represent entities, edges represent connections between them.

**Q3. Applications?**  
Social networks, recommendation systems, molecular analysis.
